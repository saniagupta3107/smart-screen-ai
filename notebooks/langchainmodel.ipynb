```python
from transformers import pipeline
from langchain_core.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate

# âœ… Load a lightweight model (good for laptops)
model_name = "google/flan-t5-base"
llm = pipeline("text2text-generation", model=model_name)

```

```python

# âœ… System Message (not strictly required for flan-t5, but included for completeness)
system_message_prompt = SystemMessagePromptTemplate.from_template(
    "You are a smart AI that extracts structured resume data from plain text."
)

# âœ… Human Message Prompt Template
human_message_prompt = HumanMessagePromptTemplate.from_template("""
Extract the following information from the resume below:

Resume:
{context}

Return the result in this structured format:

1. Contact Information:
- Name:
- Email:
- Phone Number:
- LinkedIn:
- GitHub:

2. Education:
- Institution Name:
- Degree:
- Field of Study:
- Graduation Year:

3. Experience:
- Job Title:
- Company Name:
- Duration:
- Responsibilities:

4. Projects:
- Title:
- Description:
- Technologies Used:

5. Skills:
- Programming Languages:
- Tools/Frameworks:

Question:
{question}

Extracted Information:
""")
```

```python

# âœ… Chain Function
def LLM_chain(context, question):
    template = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])
    messages = template.format_messages(context=context, question=question)
    prompt = messages[1].content
    response = llm(prompt, max_new_tokens=512)[0]['generated_text']
    return response.strip()
```

```python

# âœ… Example Resume
resume_text = """
John Doe
Email: johndoe@example.com
Phone: +1 123-456-7890
LinkedIn: linkedin.com/in/johndoe
GitHub: github.com/johndoe

Education:
- XYZ University, B.Sc. in Computer Science, 2015â€“2019

Experience:
- Software Engineer at ABC Corp (Jan 2020 â€“ Present)
  Responsibilities: Developed and maintained RESTful APIs using Flask, improved UI using React, led a team of 3 interns.

Projects:
- Resume Parser: Built an AI tool to extract resume info using Python and Transformers.
- Portfolio Website: Created a personal portfolio using Next.js and Tailwind CSS.

Skills:
- Python, JavaScript, SQL, HTML/CSS
- Flask, React, Git, Docker
"""

# âœ… Ask the model
question = "Extract key information from this resume."
result = LLM_chain(resume_text, question)

# âœ… Print the result
print("ðŸ“„ Extracted Information:\n")
print(result)
```
